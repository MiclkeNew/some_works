```python

from tkinter import *
from tkinter import messagebox
from math import *
master = Tk()
master.title('Генератор российских и украинских синусойд')
canvas = Canvas(master, width=1000, height=600, background='grey')
canvas.grid(column=0, row=1)





def mid():
    a = int(e1.get())
    b   = int(e2.get())
    fill = 'red'
    if var.get() == 0:
        fill = 'red'
    elif var.get() == 1:
        fill = 'white'
    elif var.get() == 2:
        fill = 'blue'
    elif var.get() == 3:
        fill = 'yellow'
    for i in range(a, b):
        canvas.create_line(5*i,100*sin(i)+300, 5*i+5, 100*sin(i)+300, fill= fill , width=100)


#задал варианты цветов

l1 = Label(master, text='Начало синусойды')
l1.grid(row=2, column=0, sticky='e')
l2 = Label(master, text='КОнец')
l2.grid(row=3, column=0, sticky='e')

#задал координаты синусойды

e1 = Entry(master)
e2 = Entry(master)
e1.grid(row=2, column=1)
e2.grid(row=3, column=1)
var = IntVar()
var.set(0)
redbutton = Radiobutton(master, text = 'Red', fg ='red', variable=var, value= 0)
redbutton.grid(row=4, column=1)
greenbutton = Radiobutton(master, text = 'white', fg='Black', variable=var, value= 1)
greenbutton.grid(row=5, column=1)
bluebutton = Radiobutton(master, text ='Blue', fg ='blue', variable=var, value= 2)
bluebutton.grid(row=6, column=1)
redbutton = Radiobutton(master, text = 'Yellow', fg ='Yellow', variable=var, value= 3)
redbutton.grid(row=7, column=1)
b1 = Button(master, text="Let's GO!", command = mid)
b1.grid(row=8, column = 1)


master.mainloop()
 
```
